# Data Handling Using Pandas

```python
import pandas as pd
```

```python
df=pd.read_excel("students.xlsx")
```

```python
print("All Students Data")
df
```

```python
display(df)
```

```python
top_students=df[df["Marks"] > 80]
```

```python
top_students
```

```python
rejected_students=df[df["Marks"]<=80]
```

```python
rejected_students
```

```python
top_students.to_excel("selected_students.xlsx",index=False)
print("Filtered data saved to 'selected_students.xlsx'")
```

```python
rejected_students.to_excel("rejected_students.xlsx",index=False)
print("File Created")
```

```python
rejected_students.to_excel("rejected2_students.xlsx",index=False)
print('File Created "rejected2_students.xlsx"')
```

# # Filter data of persons whose weight greater 60kg. We want only Age,Gender,Weight,Protein_per_kg

```python
select_data=data[["Age","Gender","Weight (kg)","protein_per_kg"]]
select_data
```

```python
select_data=data[data["Weight (kg)"]>60]
final_data=select_data[["Age","Gender","Weight (kg)","protein_per_kg"]]
final_data
```

```python
final_data=select_data["Weight (kg)"] > 60
final_data
```

```python
final_data.to_csv("overweight.csv", index=False)
print("Filtered data saved to 'overweight.csv'")
```
